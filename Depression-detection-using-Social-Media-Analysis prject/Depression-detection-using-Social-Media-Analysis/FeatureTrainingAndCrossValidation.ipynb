```python
X=data2.iloc[:,1:].values
y=data2.iloc[:,0].values
```

## Top 30 features after Feature Selection

```python
x=data2.iloc[:,1:]
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
best = SelectKBest(score_func=chi2, k=30)
fit = best.fit(X, y)
dfscores=pd.DataFrame(fit.scores_)
dfcolumns=pd.DataFrame(x.columns)
featureScores=pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns=['specs','scores']
featureScores
print(featureScores.nlargest(30,'scores'))
```

## Random Forest Using Top 20 Features 

```python
x_features=x[['Tone','Clout','Authentic','negemo','posemo','Analytic','sad','function','i','negate','Sixltr','verb','cogproc','Dic','anger','focuspresent','leisure','differ','social','pronoun']]
```

```python
x_features=x_features.values
x_features
```

```python
from sklearn.model_selection import train_test_split;
X_train,X_test,y_train,y_test=train_test_split(x_features,y,test_size=0.2,random_state=0)
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=9,random_state=45,min_samples_split=50)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
accuracy_score(y_test,y_pred)
```

### Applying Cross Validation 

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(clf,x_features,y,cv=10,scoring='accuracy').mean())
```

## K Nearest Neighbors Using Top 20 Features 

```python
from sklearn.model_selection import train_test_split;
X_train,X_test,y_train,y_test=train_test_split(x_features,y,test_size=0.2,random_state=0)
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
accuracy_score(y_test,y_pred)
```

## Applying Cross Validation 

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(classifier,x_features,y,cv=10,scoring='accuracy').mean())
```
